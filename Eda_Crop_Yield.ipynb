**Initial Setup & Data Loading**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/crop_yield_dataset.csv")
df.head()
```

**Basic Info**

```python
df.shape
df.info()
df.describe()
```

**Check Missing Values**

```python
df.isnull().sum()
```

**Univariate Analysis**

```python
df.hist(figsize=(10,8))
plt.show()
```

```python
sns.countplot(x='Soil Type', data=df)
plt.show()

sns.countplot(x='Weather Condition', data=df)
plt.show()

sns.countplot(x='Crop Type', data=df)
plt.xticks(rotation=45)
plt.show()
```

**Bivariate Analysis**

```python
sns.scatterplot(x='Temperature (°C)', y='Yield (tons/hectare)', data=df)
plt.show()
```

```python
sns.scatterplot(x='Rainfall (mm)', y='Yield (tons/hectare)', data=df)
plt.show()
```

```python
sns.scatterplot(x='Humidity (%)', y='Yield (tons/hectare)', data=df)
plt.show()
```

**Categorical vs Yield**

```python
sns.boxplot(x='Soil Type', y='Yield (tons/hectare)', data=df)
plt.show()

sns.boxplot(x='Weather Condition', y='Yield (tons/hectare)', data=df)
plt.show()

sns.boxplot(x='Crop Type', y='Yield (tons/hectare)', data=df)
plt.xticks(rotation=45)
plt.show()
```

**Correlation Heatmap**

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

**Group Analysis**

```python
df.groupby('Crop Type')['Yield (tons/hectare)'].mean()

df.groupby('Soil Type')['Yield (tons/hectare)'].mean()

df.groupby('Weather Condition')['Yield (tons/hectare)'].mean()
```
